```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
```

```python
np.random.seed(42)
X = np.linspace(0, 10, 50).reshape(-1, 1)
y = 2 + 3 * X - 0.5 * X**2 + np.random.normal(0, 1, X.shape)
```

```python
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
```

```python
model = LinearRegression()
model.fit(X_poly, y)
```

```python
y_pred = model.predict(X_poly)
```

```python
mse = mean_squared_error(y, y_pred)
print(f"Mean Squared Error: {mse}")
```

```python
plt.scatter(X, y, color='blue', label="Original Data")
plt.plot(X, y_pred, color='red', label="Polynomial Regression Line")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.show()
```

#Task 1: Load and Visualize the Data

```python
df=pd.read_csv('/content/Salary_Data.csv.csv')
print(df)
```

```python
np.random.seed(0)
X = np.random.rand(100, 1) * 10
y = 2 * X + 5 + np.random.randn(100, 1) * 2
df = pd.DataFrame({'Years of Experience': X.flatten(), 'Salary': y.flatten()})
```

```python
plt.scatter(df['Years of Experience'], df['Salary'])
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Relationship between Years of Experience and Salary')
plt.show()
```

# Task 2: Apply Polynomial Regression

```python
# Preprocess the data
X = df['Years of Experience'].values.reshape(-1, 1)
y = df['Salary']
```

```python
poly = PolynomialFeatures(degree=2)
X_poly = poly.fit_transform(X)
model = LinearRegression()
model.fit(X_poly, y)
```

# Task 3: Evaluate and Visualize

```python
# Make predictions
y_pred = model.predict(X_poly)
```

```python
# Evaluate the model
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

```python
plt.scatter(X, y, color='blue', label='Original Data')
plt.plot(X, y_pred, color='red', label='Polynomial Regression Curve')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Polynomial Regression')
plt.legend()
plt.show()
```
