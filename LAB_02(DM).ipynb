```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.hist(figsize=(12, 10), bins=20)
plt.suptitle('Histograms of Dataset Features')
plt.show()
```

```python
o=[1,3,5,7,9]
e=[2,4,6,8,10]
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=e, y=o)
plt.title('Scatter Plot')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x=df['age'], y=df['chol'])
plt.title('Scatter Plot')
plt.show()
```

```python
data = pd.DataFrame({
    'Math': [85, 78, 92, 88, 76],
    'Science': [88, 79, 95, 84, 82],
    'English': [75, 83, 89, 92, 80]
})

# correlation matrix
correlation_matrix = data.corr()

# heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')

plt.title('Correlation Heatmap')
plt.show()
```

```python
x1 =[0, 1, 2, 3]
y1 =[3, 8, 1, 10]
x2 =[0, 1, 2, 3]
y2 = [6, 2, 7, 11]

plt.plot(x1, y1, x2, y2)
plt.show()
```

```python
x = ["A", "B", "C", "D"]
y = [3, 8, 1, 10]

plt.bar(x, y, color = "#4CAF50")
plt.show()
```

```python
y = [35, 25, 25, 15]
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(df['age'], df['thalach'], 'b-', alpha=0.7, label='Max Heart Rate')
plt.title('Age vs. Maximum Heart Rate Achieved')
plt.xlabel('Age')
plt.ylabel('Max Heart Rate (thalach)')
plt.grid(True)
plt.legend()
plt.show()
```

```python
cp_counts = df['cp'].value_counts().sort_index()  # Change 'data' to 'df'
cp_labels = ['Type 0', 'Type 1', 'Type 2', 'Type 3']
plt.figure(figsize=(8, 5))
plt.bar(cp_labels, cp_counts, color='orange', alpha=0.8)
plt.title('Chest Pain Types Distribution')
plt.xlabel('Chest Pain Type')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
gender_counts = df['sex'].value_counts()
gender_labels = ['Male', 'Female']
plt.figure(figsize=(7, 7))
plt.pie(gender_counts, labels=gender_labels, autopct='%1.1f%%', startangle=140, colors=['skyblue', 'pink'])
plt.title('Gender Distribution')
plt.show()
```
